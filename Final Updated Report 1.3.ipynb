## 2. Mathematical Model ##
## 2-A ##
There are three different methods for the government to reduce flooding damage:
1. Construction of seawalls
2. Flood-proofing of buildings
3. Offering a subsidy to encourage people to move to a new neighbourhood

Assume there is an area (with $n$ houses) that will soon be affected by a flood, and the government can make an investment in each of the three methods mentioned above to reduce the flooding damage. The objective is for the government to minimize the sum of flooding damage and the total sum of the investments made. The first two methods will reduce a certain percentage of the damage, while the last method (offering a subsidy) will avoid all loss for the target house. The residents in a house that receives a subsidy will be assigned to a temperatory living area with finite capacity, thus the government must decide which families to move to reach the optimal cost. In this model we assume the flooding only causes loss from an economic perspective.

#### Assumptions ###
1. When a subsidy is above half the price of a given house, that family will take it and move.

2. The cost of flood-proofing a home is fixed for all houses.

3. The percentage investment in seawall construction has a linear relationship with the percentage reduction in housing damage costs.

4. The houses are choosen from a risky area in New York and we assume all houses are impacted by the same flood.

5. Each family averages 4 people.

#### Variables and Notation ### 
 
 $x_i$: Binary decision variable for the government to decide whether to assign a flood-proofing plan to house $i$.
 
 $y_i$: Binary decision variable for the government to decide whether to offer a subsidy to house $i$.
 
 $c$: Fixed cost of flood-proofing any home
 
 $S$: Amount of subsidy to offer
 
 $I$: Amount of investment in seawall construction
 
 $n$: Index for each house in the risky area
 
 $B$: Budget limit for the government's total investment
 
 $M_i$: Housing price for house $i$
 
 $u_i$: Number of residents in house $i$
 
 $N$: The capacity of the new relocation area
 
 $C_I = f(I)$: Percentage of reduced housing damage cost due to seawall construction (a function of investment)
 
 $C_FP$: Each house $i$ receiving the flood-proofing method will reduce the flooding damage by a percentage $C_FP$
 
 $L$: Total flooding cost
 
#### Constraints ### 
 
 1. A house $i$ can either accept the subsidy to move or stay and have their home flood-proofed:
 
 $$0\le x_i+y_i\le 1$$
 

 
 2. The sum total investment for the three methods should be less than the government budget:
 
 $$\sum_{i=1}^{n}cx_i+\sum_{i=1}^{n}Sy_i+I\le B$$
 
 3. The size of the subsidy must be higher than the half of the housing price to make residents want to move and the upper bound of the subsidy is the minimum housing price:
 
 $$\min(\frac{1}{2}M_i)\le S\le \min({M_i})$$

4. The total number of people to relocate should less than the capacity of new relocation area

$$\sum_{i=1}^{n}u_iy_i\le N$$

5. The total reduced flooding cost is a function of $S$, $I$, $C_I$ and $C_FP$ while the orignal cost is the sum of all housing prices

$$E[L] = \sum_{i=1}^{n}M_i - f(S, I, C_I, C_FP)$$
### Objective###

The objective is for the government to minimize both the flooding cost and the investment to reduce the damage.

$$\sum_{i=1}^{n}cx_i+\sum_{i=1}^{n}Sy_i+I+\lambda E[L]$$

### Full Model###


$$\underset{x_i, y_i, S, I}{\text{minimize}}\qquad\sum_{i=1}^{n}cx_i+\sum_{i=1}^{n}Sy_i+I+\lambda E[L] $$
  $$\qquad 0\le x_i+y_i\le 1$$
    $$ \sum_{i=1}^{n}cx_i+\sum_{i=1}^{n}Sy_i+I\le B $$
    $$ \min(\frac{1}{2}M_i)\le S \le \min({M_i}) $$
    $$ \sum_{i=1}^{n}u_iy_i\le N $$
    $$ E[L] = \sum_{i=1}^{n}M_i - f(S, I, C_I, C_FP)$$
   

 ## 2-B ##
 Using all the zip codes in Queens as our dataset, we assign each zip code to the hospital where its population would receive any necessary care in the case of a flood. 

#### Model Summary
To obtain our dataset, we averaged the x and y coordinates of all addresses in each zip code to obtain each zip code's x and y coordinate below. We recorded the sum of the addresses in each zip code to represent their approximate population. Given the location of each of the fifteen hospitals in Queens, we assigned each zip code to one hospital, making sure to keep the number of addresses assigned to each hospital relatively equal. We assigned these hospitals in a way that minimizes the total distance of each zip code to its given hospital.

#### Parameters
$\textit{H}$: The set of hospitals containing their x and y coordinate.      
$\textit{n}$: The number of hospitals in the dataset.      
$\textit{Z}$: The set of zip codes containing their x and y coordinate.       
$\textit{a}$: The number of addresses in a given zip code.        
#### Decision Variables
$A_{i,j}$: The binary assignment matrix, with a 1 representing the assignment of zip code i to hospital j.       
$d_{i,j}$: The distance from zip code i to hospital j.
#### Constraints
1. Each hospital can't be responsible for serving more than 6.875% of the population. Split evenly, they would each be responsible for 6.25% of the population.
$$
\sum_{i \in \textit{Z}} (a_{\textit{i}}*A_{\textit{i,j}})<= ((\sum_{i \in \textit{Z}} a_{\textit{i}})*1.1 ) / n \quad \forall j \in H
$$

2. Each zip code may only be assigned to a single hospital.
$$
\sum_{j \in \textit{H}} A_{i,j}=1 \quad \forall i \in Z
$$

  
#### Objective
Our objective is to minimize the distance from each zip code to its assigned hospital.

$$\underset{A_{i,j}, d_{i,j}}
{\text{minimize}} \sum_{i\in \textit{Z} j\in \textit{H}}d_{i,j} A_{i,j}
$$ 

## 2-C ##
We will solve the two problems using two different variants of a Vehicle Routing Problem formulation as discussed below. An interesting reference on VRPs with time windows constraints is found in a work by [Brian et. al](#References:) and was used to seek help while preparing the mathematical model for this scenario.
### Variant 1
#### Problem Setting
Uncapacitated vehicles with no demand or time constraints.

First, we solve a general variant of a Vehicle Routing Problem with no capacity constraints on each vehicle and no time windows for delivery at customer locations. All vehicles need to return to their starting points and all vehicles must be used. We ignore the service time necessary for each customer location in this variant. We model this problem using nodes and arcs. We will consider two indexed variables in this model contrary to the second variant where we will consider three. We use Miller-Tucker-Zemlin method in this formulation to eliminate subtours (this method is not used in our second formulation).

#### Parameters
$\textit{N}$: This is a set of nodes. Each node consists of x and y coordinates. 


$\textit{C}$: The set of customer nodes which doesn't include the depot nodes.

$\textit{A}$: Set of all routes possible between all pairs of nodes (a tuple of two nodes).

$d_{i,j}$: Distance between node i and node j.
#### Decision Variables
$u_i$: A variable representing the position of node i on the route.

$x_{i,j}$: A binary decision variable to account for whether node i and node j are directly connected.
#### Constraints
1. Each vehicle should start from the depot. Thus the total number of arcs exiting the depot is equal to the total number of vehicles.
$$
\sum_{j \in \textit{N}} x_{\textit{depot},j}=v
$$

2. All vehicles should return to the depot. Thus the total number of arcs entering the depot is equal to the total number of vehicles.
$$
\sum_{i \in \textit{N}} x_{i,\textit{depot}}=v
$$

3. Each location is visted by exactly one vehicle i.e. there is an incoming arc at each location.
$$
\sum_{i \in \textit{N}} x_{i,j}=1 \quad \forall j \in \textit{C}
$$

4. Exactly one vehicle leaves from each location i.e. there is an outgoing arc at each location.
$$
\sum_{j \in \textit{N}} x_{i,j}=1 \quad \forall i \in \textit{C}
$$

5. A vehicle cannot travel the same arc twice i.e. it cannot go from i to j on $arc_{ij}$ and return from j to i on the same arc.  
$$
x_{ij}+x_{ji}\le 1 \quad \forall i \in \textit{N} \quad \forall j \in \textit{C}
$$

6. Self nodes cannot be connected. A vehicle cannot travel from node i to node i.
$$
x_{ii}=0 \quad \forall i \in \textit{N}
$$

7. Subroutine elimination - we use a standard approach of subroutine elimination in this method (Miller-Tucker-Zemlin).
$$
u_i-u_j+v*x_{i,j}\le v-1
$$
  
  
#### Objective
Our aim is to minimize the distance traveled by the vehicles. Thus our objective is a direct function of the active arcs and length of those arcs.
$$\underset{u_i, x_{i,j}}
{\text{minimize}} \sum_{i\in \textit{N} j\in \textit{N}}d_{ij}*x_{ij}
$$

### Variant 2

#### Problem Setting

Capacitated vehicles with demand constraints.

We consider that the government has a maximum of v vehicles available and not all vehicles need to be used. Each vehicle starts its trip from the depot and ends the trip at a given destination i.e. it doesn't come back to the depot. We also take into account the unloading time at each location. Each customer has a given demand, and no vehicle can service more customers than their capacity permits. Thus we summarize the complex constraints for this VRP below:

1. Time windows for delivery: Each customer location has a desired window of time during the day to accept the delivery.
2. Service time at each location: Each location has a different service time depending on the amount of materials to be unloaded.
3. Capacitated vehicles: All vehicles are identical and have a maximum capacity. 
4. DIfferent start and destination: The tour starting and ending points are different, thus the vehicles route is not a closed loop.

The goal is to design a set of routes that minimizes total cost, such that:

1. Each customer is served exactly once.
2. Every route originates at the depot and ends at specified common node.
3. Time windows for each customers delivery and the capacity constraints of the vehicles are observed. 

#### Parameters
$\textit{N}$: A set of nodes, each consisting of x and y coordinates  
$\textit{C}$: Set of nodes consisting of all customers, which doesn't include the depot and the destination  
$\textit{A}$: Set of all possible routes between each pairs of nodes. It is a tuple of two nodes  
$d_{i,j}$:  Distance between node i and node j  
$[a_{i},b_{i}]$: Each customers time window, starting after $a_i$ until $b_i$  
$t_i$: Service time required at each customer location  
$q$: Maximum capacity of each vehicle  
$d_i$: Demand requirement at each customer location  
$v$: Set of all vehicles available, all considered to be identical.

#### Decision Variables
$x_{ijk}$: A binary decision variable that takes value 1 if the arc connecting node i and node j is commuted by vehicle k.  
$s_{ik}$: This decision variable is defined for each vertex i and each vehicle k and denotes the time vehicle k starts to service customer i. 

#### Constraints
1. All vehicles start at time zero from the depot. 
$$
s_{depot,k}=0\quad \forall k \in v
$$
2. All vehicles start from the depot.
$$
\sum_{j \in \textit{N}}x_{depot, j,k}=1 \quad \forall k \in v
$$

3. Destination node has no outgoing vehicles or no arcs originate from the destination node.
$$
\sum_{j \in \textit{N}}x_{destination, j,k}=0 \quad \forall k \in v
$$

4. Depot node has no incoming vehicles or no arcs terminate at the depot node.
$$
\sum_{i \in \textit{N}}x_{i, depot,k}=0 \quad \forall k \in v
$$

5. A vehicle $k$ cannot travel on the same arc twice i.e. it cannot go from i to j on $arc_{ij}$ and return from j to i on the same arc.  
$$
x_{ijk}+x_{jik}\le 1 \quad \forall i \in \textit{N} \quad \forall j \in \textit{C} \quad k \in v
$$

6. A node cannot be connected to itself.
$$
x_{i,i,k}=0 \quad \forall i \in \textit{N} \quad \forall k \in v
$$

7. Capacity constraint - the cumulative demand of the locations on a vehicle's route must be less than the vehicle's capacity.
$$
\sum_{i \in \textit{C}}d_i*[\sum_{j \in \textit{N}} x_{ijk}]\le q \quad \quad \forall k \in v
$$

8. After a vehicle has arrived at a customer location, it should leave for another destination.
$$
\sum_{i \in \textit{N}} x_{ihk}-\sum_{j \in \textit{N}} x_{hjk}=0 \quad \forall h \in \textit{C} \quad \forall k \in v  
$$

9. Relationship between customer departure time from one customer and its immediate successor.
$$
x_{ijk}*[s_{ik}+t_i-s_{jk}\le 0]\quad \forall i \in \textit{N} \quad \forall j \in \textit{N} \quad \forall k \in v  
$$
10. We ensure that each customer is served in their desired time window.
$$
a_i \le s_{ik}\le b_i
$$

It should be noted here that we don't need a subroutine elimination in this formulation as it has already been taken care of implicitly. The service start variables impose a unique route direction thereby eliminating any subtours. Hence, the classical VRP subtour elimination constraints become redundant in this formulation.

#### Objective
Our aim is to minimize the total distance traveled by the vehicles while satisfying all the problems requirements:
$$\underset{x_{ijk}, s_{ik}}
{\text{minimize}} \sum_{a \in \textit{A} k \in v}d_i*x_{ak}
$$

## 3. Solution ##
### 3(A) ###
First step of Part A is to find the plot fit for seawall investment and percentage reducing cost. Seawall construction will reduce the flooding damage from 10% to 50%. This value is positive related to the investment. The following code finds the relations between I and cost reducing percentage:

```julia
# seawall construction cost
# we assume that the lenth of seawall to cover all the houses will be at least 650 feet
x1 = [422500, 494000, 559000, 861900, 958750, 1157000, 125000, 1300000] 
y1 = [0.1, 0.17, 0.23, 0.30, 0.39, 0.45, 0.48, 0.5]

# order of polynomial to use
k = 1
n = length(x1)
A = zeros(n,k+1)
for i = 1:n
    for j = 1:k+1
        A[i,j] = x1[i]^(k+1-j)
    end
end

using JuMP, Gurobi
m = Model(solver=GurobiSolver(OutputFlag=0,BarHomogeneous=1))

@variable(m, u[1:k+1])
@objective(m, Min, sum( (y1 - A*u).^2 ) )
status = solve(m)
uopt = getvalue(u)
println(status)
println(getobjectivevalue(m))
println(uopt)
findmin(x1)
```

The second step is to figure out the optimal strategy for governmnet to invest. 

```julia
#Input housing price

M = [501300 514840 423000 371700 289900 343350 348300 379230 279800 312500]

area = [1870 2480 1800 2120 1864 1416 1392 1600 1732 1092]

#flood proofing cost
I = 75*area #The cost for flood-proofing is calculated by 75 times the area of house. 


HouseID = [1:10]
num_families = [1 2 1 2 2 2 1 1 1 1]
#housing price denotes as M

using JuMP, Gurobi, ConditionalJuMP
m=Model(solver=GurobiSolver(OutputFlag=0, NumericFocus=3, BarHomogeneous=1))

# Number of people in each house
P = 5*num_families

t = length(M)

a = 0.4 #flood proofing parameter (cost reducing percentage)

@variable(m, 1>=y[1:t]>=0, Bin) #subsidy decision variable
@variable(m, 1>=x[1:t]>=0, Bin) #flood-proofing decision variable
@variable(m, z>=422500) #Investment for seawall construction
@variable(m, 289900>=s>=0) #value of subsidy


reduced_cost_S = (M*y)[1]
reduced_cost_FP = (a*M*x)[1]
z_input = [z 1]
reduced_cost_SC = (sum(M)*(z_input*uopt))[1]
#each house can only accept subsidy to move or get the flood proof
for i = 1:t
    @constraint(m, x[i]+y[i]==1)
end

C = findmin(M)
c = C[:1] #upper bound for subsidy (minimum house value)
c_2 = 0.5*c #lower bound for subsidy (half of the minimun house value)


#if subsidy less than the half of housing value peopel will stay
for i =1:t
    @implies(m, (s<=0.5*M[i]) => (y[i]==0)) 
end
sum_s = 0
for i = 1:t
    sum_s += s*y[i]
end

B = 0.5*sum(M)# the total investment less than half of the total housing price

lamda = 2 #weight parameter

@constraint(m, P*y.<=50) #capcity for new relocation place
@constraint(m, sum_s+z+I*x.<=B) 
obj = lamda*sum(A)-reduced_cost_S-reduced_cost_FP-reduced_cost_SC

@objective(m, Min, obj)

state = solve(m)

cost_fp = 0
for i = 1:10
    cost_fp += getvalue(x[i])*I[i]
end

println("Decision for flood-proofing is ",getvalue(x))

println("Decision for offerring subsidy is ",getvalue(y))
```

```julia
data = readcsv("queens_MapPLUTO_dataset_noOutliers.csv")
# went from 65 to 56 data points (removed 9 outliers)

# data in order: zip code, x coord, y coord, number addresses in zip code
zip_codes = data[:,1]
coords = data[:,2:3] # average of all x and y coords in each zip codes
addresses = data[:,4] # represent population
num_zips = length(zip_codes)
num_addresses = sum(addresses)

# transformed hospital coordinates
X = [1.02192e6 2.15792e5; 
    1.01637e6 1.71076e5; 
    1.02554e6 2.08366e5; 
    1.05344e6 216686.0; 
    1.00325e6 2.09124e5; 
    1.01195e6 2.11023e5; 
    1.00282e6 204289.0; 
    1.01476e6 2.17358e5; 
    1.02665e6 2.0709e5; 
    1.03205e6 2.0498e5; 
    1.0427e6 2.1219e5; 
    976648.0 1.96462e5; 
    9.94999e5 194067.0; 
    1.00896e6 191684.0; 
    1.00314e6 1.85691e5]

# number of hospitals
n = 15

# each hospital can't be responsible more than 6.875% of the population
# split evenly, they would each be responsible for 6.25%
SUPPLY_CONSTRAINT_HOSPITAL = (num_addresses*1.1)/n;
```

```julia
using JuMP, Gurobi
m = Model(solver=GurobiSolver(OutputFlag=0))

@variable(m, A[1:n, 1:num_zips], Bin) # binary assignment matrix

for j=1:num_zips
  @constraint(m, sum( A[:,j] ) == 1 ) # each zip code assigned to one hospital
end

# no hospital can be responsible for too much of the population
for i=1:n
  @constraint(m, sum( A[i,j]*addresses[j] for j=1:num_zips) <= SUPPLY_CONSTRAINT_HOSPITAL)
end

@expression(m, distance_to_hospital, sum( norm(coords[i,:] - X[j,:]) * A[j,i] for i=1:num_zips, j=1:n) )

# minimize x and y distance from each zip code to its assigned base
@objective(m, Min, distance_to_hospital )

solve(m)
aopt = getvalue(A);
```

## 3-C Variant 1 ##

```julia
#Forming Arcs and Nodes using X,Y coordinates of Depots and Customer Locations
            
arcs=[] #Defining a blank List to store arcs
nodes=[] #Defining a blank list to store nodes
new=data; #Storng the data matrix consisting of X and Y coordinates in a new variable

for i=1:length(new[:,1])
    a=new[i,:]
    push!(nodes,(a[1],a[2])) #Appending the list of nodes           
    for j=1:length(new[:,1])
        b=new[j,:];
            push!(arcs,((a[1],a[2]),(b[1],b[2]))) #Appending the list of arcs
    end
end

#Calculating distance/length of each arc 
dist=Dict();  #Creating a dictionary to store length of each arc
for i=1:length(arcs)
    arc=arcs[i];
    d=((arc[1][1]-arc[2][1])^2+(arc[1][2]-arc[2][2])^2)^0.5
    s=string(arcs[i])
    push!(dist,s=>d)
end
```

```julia
using JuMP, Gurobi, PyPlot
m=Model(solver=GurobiSolver(OutputFlag=0))
l=length(data[:,1])  #number of points to be covered
n=length(arcs) #Number of arcs
v=4; #Number of vehicles

@variable(m,u[nodes]) #position of node i
@variable(m,x[arcs],Bin)    #Arc ij connected by vehicle k

@constraint(m,depot_leave,sum(x[(nodes[1],j)] for j in nodes)==v) #Exactly v vehicle depart from depot
@constraint(m,depot_return,sum(x[(i,nodes[1])] for i in nodes)==v) #Exactly v vehicle return to depot

#One tour leaves each node
for i in nodes[2:16]
    @constraint(m,sum(x[(i,j)] for j in nodes)==1)                
end

#One tour enters each node
for j in nodes[2:16]
    @constraint(m,sum(x[(i,j)] for i in nodes)==1)                
end

#Only one in ij and ji is included-for i in nodes
for i in nodes[2:16]
    for j in nodes
        @constraint(m,x[(i,j)]+x[(j,i)]<=1)
    end 
end


#Not self
for i in nodes[1:16]
    for j in nodes
        if i==j
            @constraint(m, x[(i,j)] == 0)
        end
    end
end

# Subroutine Elimination
for i in nodes[2:16]
    for j in nodes[2:16]
        if i!=j
             @constraint(m,u[i]-u[j]+v*x[(i,j)]<=v-1)
        end 
    end
end

#Minimizing the distance travelled
@objective(m,Min,sum(((i[1][1]-i[2][1])^2+(i[1][2]-i[2][2])^2)^0.5*x[i] for i in arcs))

solve(m)
a=getvalue(x);
#################################
```

## 3-C-Variant-2 ##

```julia
# Data
# Converting list ot arcs
data = [1.06143e6 1.90554e5;
    1.02192e6 2.15792e5; 
    1.01637e6 1.71076e5; 
    1.02554e6 2.08366e5; 
    1.05344e6 216686.0; 
    1.00325e6 2.09124e5; 
    1.01195e6 2.11023e5; 
    1.00282e6 204289.0; 
    1.01476e6 2.17358e5; 
    1.02665e6 2.0709e5; 
    1.03205e6 2.0498e5; 
    1.0427e6 2.1219e5; 
    976648.0 1.96462e5; 
    9.94999e5 194067.0; 
    1.00896e6 191684.0; 
    1.00314e6 1.85691e5] #First Point is starting point, Last point is ending Point

arcs=[];
new=data;
nodes=[]
for i=1:length(new[:,1])
    a=new[i,:]
    push!(nodes,(a[1],a[2])) #Storing all nodes
    for j=1:length(new[:,1])
        b=new[j,:];
        push!(arcs,((a[1],a[2]),(b[1],b[2])))  #Storing all possible arcs
    end
end
customers=nodes[2:15]

# calculating distance of each arc
dist=Dict();
for i=1:length(arcs)
    arc=arcs[i];
    d=((arc[1][1]-arc[2][1])^2+(arc[1][2]-arc[2][2])^2)^0.5
    s=string(arcs[i])
    push!(dist,s=>d)
end

############
v=1:5 #Number of vehicles available
q=130000; #Capacity of each vehicle
d=[20897,21134,17815,12939,19874,19301,22151,19571,20184,21682,21661,20305,21516,22098,21078] # Used from result of Part b

demand=Dict(zip(customers,d))

#We have 15 customers
#Time zero is equivalent to 7am; To represent window 10am-12 am--We say time=>3-5
a=[0,2,0,1,0,4,0,2,0,6,0,1,0,8,10,0]; #Starting indow time, randomly generated
a=Dict(zip(nodes,a)) #Dictionary for opening time of delivery

b=[15,12,10,8,14,14,14,10,14,12,14,14,12,14,14,14]; #Ending Window time, randomly generated

b=Dict(zip(nodes,b)); #Dictionary for closing time of delivery
t=[1,2,1.5,2.5,3,1,2,2.5,2,3,1,2,1,1.5,3,3]; #Service time required at each customer location
t=Dict(zip(nodes,t));
```

```julia
using JuMP, Gurobi
m=Model(solver=GurobiSolver(OutputFlag=0))

@variable(m,x[arcs,v],Bin) #arc covered by vehicle k
@variable(m,s[nodes,v]) # time at which vehicle k starts to service customer i. 
@variable(m,u[nodes])

@constraint(m,depot[k in v],s[nodes[1],k]==0)  #Vehicle starts at time zero from depot
@constraint(m,depot2[k in v],sum(x[(nodes[1],j),k] for j in nodes)==1)  #Vehicle starts at time zero from depot

###
@constraint(m,last[k in v],sum(x[(nodes[16],j),k] for j in nodes)==0) #Vehicle should not leave from destination node
@constraint(m,first_last[k in v],sum(x[(i,nodes[1]),k] for i in nodes)==0) #Last node on route cannot be Depot

@constraint(m,diffroues[k in v, i in nodes, j in nodes],sum(x[(i,j),k]+x[(j,i),k])<=1) #Same route is not traversed twice

@constraint(m,self[k in v],sum(x[(i,i),k] for i in nodes)==0) #No self arcs

@constraint(m,single[i in customers],sum(x[(i,j),k] for j in nodes for k in v)==1) #Each customer location has 1 incoming vehicle

@constraint(m,capacity[k in v],sum(demand[i]*sum(x[(i,j),k] for j in nodes)for i in customers)<=q) #Capacity limits of vehicles
                                
@constraint(m,con2[h in customers,k in v],sum(x[(i,h),k] for i in nodes)-sum(x[(h,j),k] for j in nodes)==0) 
# a vehicle arrives at a customer has to leave for another destination; 
                
@constraint(m,relation[i in nodes, j in nodes,k in v],x[(i,j),k]*(s[i,k]+t[i]-s[j,k])<=0)
#relationship between the vehicle departure time from a customer and its immediate successor

#time windows are observed
@constraint(m,integrality1[i in nodes,k in v],s[i,k]<=b[i])
@constraint(m,integrality2[i in nodes,k in v],s[i,k]>=a[i])

#Minimizing the distance
@objective(m,Min,sum(dist[string(a)]*x[a,k]for a in arcs for k in v))
solve(m)
```

## Further Improvements

Our Vehicle Routing Model considers that all vehicles are identical and cost the same. An improvement could be utilizing distinct vehicles with different capacities and demands. Moreover, we have considered only single depot for our problem. If the vehicles are based at different depots, one subproblem must be solved for each depot. A good assumption would be that the vehicles can finish their routes at a depot different from where they started, but the number of vehicles starting and ending at a depot remains constant. 

Another doable but tough extension would be to consider stochastic demand at each customer location.
